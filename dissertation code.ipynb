# feature selection

```python
from sklearn.feature_selection import VarianceThreshold

# Create a VarianceThreshold instance for feature selection
var_thr = VarianceThreshold(threshold=0.0)

# Fit the VarianceThreshold instance on the training features
var_thr.fit(X_train)
```

```python
# Get the number of selected features after applying VarianceThreshold
selected_features_count = var_thr.get_support().sum()

# Print the result
print(f"Number of selected features: {selected_features_count}")
```

```python
# Get the selected columns after applying VarianceThreshold
selected_columns = X_train.columns[var_thr.get_support()]

# Get the columns that were not selected
not_selected_columns = X_train.columns[~var_thr.get_support()]

# Print the list of not selected columns
print("Columns not selected by VarianceThreshold:")
print(not_selected_columns)
```

```python
# Calculate the Pearson correlation coefficient matrix
corr = X_train.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure and set its size
plt.figure(figsize=(20, 10))

# Generate a heatmap using the correlation matrix
heatmap = sns.heatmap(corr)

# Show the heatmap
plt.show()
```

```python
def find_highly_correlated_columns(dataset, threshold):
    correlated_columns = set()
    correlation_matrix = dataset.corr()

    for i, col_i in enumerate(correlation_matrix.columns):
        for j, col_j in enumerate(correlation_matrix.columns[:i]):
            correlation_value = correlation_matrix.iloc[i, j]
            
            if abs(correlation_value) > threshold:
                correlated_columns.add(col_i)
                correlated_columns.add(col_j)

    return correlated_columns
```

```python
corr_features=find_highly_correlated_columns(X_train,0.99)
len(set(corr_features))  
```

```python
# Removing highly correlated features from X_train and X_test
X_train = X_train.drop(columns=corr_features)
X_test = X_test.drop(columns=corr_features)
```
